```python
import chipwhisperer as cw
import matplotlib.pyplot as plt

n = 4

scope = cw.scope()
target = cw.target(scope)

scope.default_setup()

cw.program_target(scope, cw.programmers.STM32FProgrammer, "./recall-CWNANO.hex")

scope.arm()

# scope.adc.samples = 800
scope.adc.samples = 1000
```

```python
import time
def reset_target(scope): 
    scope.io.nrst = 'low'
    time.sleep(0.01)
    scope.io.nrst = 'high'
    time.sleep(0.01)

def cap_pass_trace(pass_guess):
    ret = ""
    reset_target(scope)
    num_char = target.in_waiting()
    while num_char > 0:
        ret += target.read(num_char, 16)
        time.sleep(0.01)
        num_char = target.in_waiting()

    scope.arm()
#     target.write(pass_guess)
    target.simpleserial_write('a', pass_guess)

    ret = scope.capture()
    if ret:
        print('Timeout happened during acquisition')

    trace = scope.get_last_trace()
    return trace
```

```python
import numpy as np

found_elts = []

for pos in range(16):
    ref_msg = bytearray(found_elts)  + bytearray([128]*(16-len(found_elts)))
    ref_trace = cap_pass_trace(ref_msg)
    
    
    temp =[]
    for i in range(0,256):
        msg = bytearray(found_elts) + bytearray([i]) + bytearray([255] * (15-len(found_elts)) )

        trace = cap_pass_trace(msg)
#         diff = int(round(np.sum(np.abs(trace - ref_trace)), -1))
        diff = int(round(np.sum(np.abs(trace - ref_trace))))
        temp.append(diff)
        print(f"pos: {pos}, i: {i}, diff: {diff}")
        
    temp = np.array(temp)
    mean = temp.mean()
    temp = abs(temp - mean)
    max_elt = temp.argmax()
        
    print(f"position: {pos}, found_elt: {max_elt}")
    found_elts.append(max_elt)
    
#     res[pos] = temp


# found_elts = [112,48,119,51,82,49,115,107,110,48,119,108,51,100,103,51]
# ref_msg = bytearray(found_elts)  + bytearray([255]*(16-len(found_elts)))
# ref_trace = cap_pass_trace(ref_msg)

# for i in range(0,256):
    
#     msg = bytearray(found_elts) + bytearray([i]) + bytearray([255]*(15-len(found_elts)))

#     trace = cap_pass_trace(msg)
#     diff = np.sum(np.abs(trace - ref_trace))

#     print(f"i = {i} diff = {round(diff)}")

```
